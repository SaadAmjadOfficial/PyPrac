```python
# Q1
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /, **): ")
num2 = float(input("Enter second number: "))

if operator == '+':
    print(num1 + num2)
elif operator == '-':
    print(num1 - num2)
elif operator == '*':
    print(num1 * num2)
elif operator == '/':
    print(num1 / num2)
elif operator == '**':
    print(num1 ** num2)
else:
    print("Invalid operator!")
```

```python
# Q2
myList = ["apple", 42, "orange", 3.14]

for value in myList:
    if isinstance(value, (int, float)):
        print(f"Numeric value found: {value}")
```

```python
# Q4
data = {"a": 100, "b": "text", "c": 200, "d": 50}
total = 0

for val in data.values():
    if isinstance(val, (int, float)):
        total += val

print("Sum of numeric items:", total)
```

```python
# Q5
myList = [1, 2, 3, 2, 4, 5, 1, 6]
duplicates = []
seen = []

for x in myList:
    if x in seen:
        duplicates.append(x)
    else:
        seen.append(x)

print("Duplicate values:", duplicates)
```
